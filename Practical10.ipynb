```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.drop('Id',axis=1)
```

```python
numerical = []
nominal = []
for col in df.columns:
    if df[col].dtypes != 'object':
        numerical.append(col)
    else:
        nominal.append(col)

print(nominal)
print(numerical)
```

```python
df.hist(figsize=(10,8))
```

```python
sns.boxplot(data=df,orient='h')
```

```python
#Sepal Width in cm has outliers
```

```python
q1 = df['SepalWidthCm'].quantile(0.25)
q3 = df['SepalWidthCm'].quantile(0.75)
IQR = q3 - q1
low = q1 - (1.5*IQR)
high = q3 + (1.5*IQR)
df.loc[df['SepalWidthCm']<low,'SepalWidthCm'] = low
df.loc[df['SepalWidthCm']>high,'SepalWidthCm'] = high
```

```python
sns.boxplot(data=df,orient='h')
```
